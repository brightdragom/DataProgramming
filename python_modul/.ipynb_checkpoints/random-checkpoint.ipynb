```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 10
y = np.random.randn(n)
```

```python
plt.plot(range(n),y)
```

```python
plt.xlabel('Time')
plt.ylabel('Random Value')
```
