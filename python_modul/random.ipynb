```python
n = 10000
x = np.random.rand(n)
y = np.random.rand(n)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y, s=1)
xx = np.linspace(0,1,100)
plt.plot(xx, (1-xx*xx)**0.5, c='r')

pi = ((x**2 + y**2) < 1).mean() * 4
pi
```
