```python
f = lambda x: x.max() - x.min()

df = pd.DataFrame(np.arange(12).reshape(4,3),
                  columns=['A','B','C'], index = ['a','b','c','d']) 
# --> arange(12) : 0~11을 reshape(4,3) 4행 3열로 만들어라 칼럼은 A,B,C 이고 인덱스는 a,b,c,d다.


df
```

```python
df.apply(f)
```

```python
df.apply(f, axis=1) #axis = 1로 설정 > 행 별 최대값 - 최솟값 출력
```
